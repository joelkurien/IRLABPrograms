```python
df = pd.read_csv('gutenberg_data.csv')
```

```python
content = df.Text.fillna('')
corpus = []
for i in range(50):
  corpus.append(content[i])
```

```python
tfidf_vector = TfidfVectorizer(stop_words='english')
#doc1 tfidf vector= 0c + 0.53a + 0i + 2.46b
#doc2
```

```python
X=tfidf_vector.fit_transform(corpus)
```

```python
feature_names = tfidf_vector.get_feature_names()
```

```python
dense = X.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)

df.head()
```

```python
from collections import defaultdict
term_doc = {}
for i, features in enumerate(tfidf_vector.get_feature_names()):
  if features in term_doc:
    if i not in term_doc[features]:
      term_doc[features].append(i)
  else:
    term_doc[features] = [i]
  
print(term_doc)
```
